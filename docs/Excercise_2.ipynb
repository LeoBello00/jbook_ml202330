```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from tqdm import tqdm

# Convert dataframe values to float
df_values = df_Wind.astype(float)

# Calculate VIF for each variable
vif = pd.DataFrame()
vif["Variable"] = df_Wind.columns
vif["VIF"] = [variance_inflation_factor(df_values.values, i) for i in tqdm(range(df_values.shape[1]))]

# Filter variables with VIF >= 5
high_vif_variables = vif[vif["VIF"] >= 5]["Variable"].tolist()

# Print variables with high VIF
print("Variables with VIF >= 5:")
print(high_vif_variables)
```

```python
from sklearn.impute import SimpleImputer

numeric_columns = df_Fraud.select_dtypes(include='number').columns
numeric_df = df_Fraud[numeric_columns]
object_df = df_Fraud[object_columns]

imputer = SimpleImputer(strategy='median')
imputed_numeric_df = pd.DataFrame(imputer.fit_transform(numeric_df), columns=numeric_columns)

imputer = SimpleImputer(strategy='most_frequent')
imputed_object_df = pd.DataFrame(imputer.fit_transform(object_df), columns=object_columns)

df_fraud_imputed = df_Fraud.copy()
df_fraud_imputed[numeric_columns] = imputed_numeric_df
df_fraud_imputed[object_columns] = imputed_object_df

df_fraud_imputed.reset_index(drop=True, inplace=True)
```

```python
df_fraud_imputed.isnull().sum().sum()
```

```python
df_fraud_imputed.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
df_encoded = pd.get_dummies(df_fraud_imputed, columns=object_columns)
# Get the categorical columns
categorical_columns = df_fraud_imputed.select_dtypes(include='object').columns

# Create an instance of OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform the categorical columns
encoded_features = encoder.fit_transform(df_fraud_imputed[categorical_columns])

# Create a new dataframe with the encoded features
df_tmp1 = df_fraud_imputed.copy()
df_encoded = pd.concat([df_tmp1.drop(categorical_columns, axis=1), pd.DataFrame(encoded_features.toarray())], axis=1)
```
